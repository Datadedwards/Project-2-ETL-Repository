```python
import pandas as pd
from IPython.display import display, HTML
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_United_States_cities_by_crime_rate'
```

```python
tables = pd.read_html(url)
tables
```

```python
df = tables[0]
df.columns = ['State', 'City', 'Population', 'Total Violent Crime','Murder', 'Rape', 'Robbery', 'Aggravated Assult', 'Total Property Crime', 'Burglary', 'Larceny-Theft', 'Motor Vehicle Theft', 'Arson']
df.head()
```

```python
html_crime_table = df.to_html()
html_crime_table
```
